### 2. 각 연도별 최대 검거율을 가진 범죄유형을 찾아서 해당 연도 및 유형의 검거건수들의 총합을 구하시오.(검거율 = 검거건수 / 발생건수)

````python
import pandas as pd

df1_2 = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_1_2.csv")
df1_2.head()
````

````python
df1_2.info()
````

````python
# 1. 데이터프레임을 연도, 발생건수, 검거 건수, 검거율로 다시 그룹화 해야한다.
df1_2_generation = df1_2[df1_2["구분"] == "발생건수"]
df1_2_generation.head()
````

````python
df1_2_arrest = df1_2[df1_2["구분"] == "검거건수"]
df1_2_arrest.head()
````

````python
crim_group = ["범죄유형1", "범죄유형2", "범죄유형3", "범죄유형4", "범죄유형5", "범죄유형6", "범죄유형7", "범죄유형8", "범죄유형9", "범죄유형10"]
df1_2_generation["총 발생건수"] = df1_2_generation[crim_group].sum(axis=1)
df1_2_arrest["총 검거건수"] = df1_2_arrest[crim_group].sum(axis=1)
df1_2_generation.head()
````

````python
df1_2_arrest.head()
````

````python
df1_2_final = pd.merge(df1_2_generation[["연도", "총 발생건수"]], df1_2_arrest[["연도", "총 검거건수"]], on="연도")
df1_2_final.head()
````

````python
df1_2_final["검거율"] = df1_2_final["총 검거건수"] / df1_2_final["총 발생건수"]
df1_2_final.head()
````

````python
df1_2_final.sort_values("검거율", ascending=False)
# 1. 검거율이 가장 높은 연도는? 2018년
# 2. 검거율이 가장 높은 연도의 검거건수: 4590건
````

## 작업형 제2유형

※ 모델 성능 지표: Macro F1 Score

※ 타깃(라벨): 농업 유형(0,1,2의 세 가지 클래스)

````python
import pandas as pd

df2_train = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_2_train.csv")
df2_train.head()
````

````python
df2_test = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_2_test.csv")
df2_test.head()
````

````python
df2_train.info()
````

````python
df2_test.info()
````

````python
# 지역, 등급에 대한 원-핫 인코딩이 필요해보인다.

df2_train["지역"].value_counts()
````

````python
df2_train["등급"].value_counts()
````

불필요 칼럼이 늘어날 듯하여 두 칼럼 모두 매핑을 통해 숫자로 처리할 예정이다.

````python
local_dict = {
    '부산':0, '대전':1, '울산':2, 
    '충남':3, '경남':4, '인천':5, 
    '대구':6, '경기':7, '전북':8, 
    '서울':9, '광주':10,'경북':11, 
    '전남':12, '강원':13, '충북':14
       }
grade_dict = {
    'A':0, 'B':1, 'C':2}
````

````python
df2_train["지역"] = df2_train["지역"].map(local_dict)
df2_train["등급"] = df2_train["등급"].map(grade_dict)
df2_train.head()
````

````python
# test 데이터에도 동일하게 적용해야한다.
df2_test["지역"] = df2_test["지역"].map(local_dict)
df2_test["등급"] = df2_test["등급"].map(grade_dict)
````

````python
df2_train.describe()
````

````python
df2_train["라벨"].value_counts()
````

특정유형(라벨 0)이 절대 다수인 것을 확인했다. 라벨별로 데이터셋을 나눌 필요가 있을 것 같다. 혹은 클래스 가중치를 이용하면 좋을 것 같다.

````python
df2_train_0 = df2_train[df2_train["라벨"] == 0]
df2_train_1 = df2_train[df2_train["라벨"] == 1]
df2_train_2 = df2_train[df2_train["라벨"] == 2]
df2_train_0
````

````python
df2_train_1
````

````python
df2_train_2
````

````python
# 학습, 평가 데이터셋 분할
from sklearn.model_selection import train_test_split

X_0 = df2_train_0.drop("라벨", axis=1)
y_0 = df2_train_0["라벨"]

X_0_train, X_0_test, y_0_train, y_0_test = train_test_split(X_0, y_0, test_size=0.2, random_state=42)

X_1 = df2_train_1.drop("라벨", axis=1)
y_1 = df2_train_1["라벨"]
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X_1, y_1, test_size=0.2, random_state=42)

X_2 = df2_train_2.drop("라벨", axis=1)
y_2 = df2_train_2["라벨"]
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X_2, y_2, test_size=0.2, random_state=42)

X_train = pd.concat([X_0_train, X_1_train, X_2_train], axis=0)
y_train = pd.concat([y_0_train, y_1_train, y_2_train], axis=0)
X_test = pd.concat([X_0_test, X_1_test, X_2_test], axis=0)
y_test = pd.concat([y_0_test, y_1_test, y_2_test], axis=0)
````

````python
# 모델 선언(랜덤 포레스트)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
# 모델 학습
model.fit(X_train, y_train)

# 모델 평가(F1 score)
from sklearn.metrics import f1_score
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred, average="weighted")
print("F1 Score:", f1)
````

````python
# 모델 예측
X_pred = df2_test.drop("라벨", axis=1)

y_test_pred = model.predict(X_pred)

# 예측결과 저장
submission = pd.DataFrame({
    "ID": df2_test["ID"],
    "라벨": y_test_pred
})

submission.to_csv("./result/result_9th.csv", index=False)
````

### 1. 한 제조 회사에서 생산성을 높이고자 직원들의 주요 생산성 요인을 분석하기로 결정하였다. 이를 위해 200명의 직원 데이터를 수집했으며, 직원들의 근무 기간, 특성 정보, 그리고 개인적인 속성을 조사하였다.

1. design을 예측하는 다중회귀 분석을 시행한 후 유의하지 않은 설명변수 개수를 구하시오(단, 불필요한 칼럼은 제외하며, 모델의 절편향은 포함)
2. 훈련 데이터(학습용 데이터)의 예측값과 실제값의 피어슨 상관계수를 구하시오(소수점 셋째 자리에서 반올림)
3. 적합한 모델을 활용하여 테스트 데이터에서의 RMSE를 구하시오.

````python
import pandas as pd

df3_1 = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_3_1.csv")
df3_1.head()
````

````python
df3_1.info()
````

````python
df3_1.describe()
````

````python
# 시험 조건에 따라 데이터셋을 분할
df3_1_train = df3_1.iloc[:140, :]
df3_1_test = df3_1.iloc[140:, :]
````

````python
df3_1_train.shape
````

````python
df3_1_train.tail()
````

````python
# 독립 변수와 종속 변수 설정
X_train = df3_1_train.drop("design", axis=1)
y_train = df3_1_train["design"]
````

````python
# 모델 회귀식 계수 유의성 검정(몰랐음)
import statsmodels.api as sm
X_train_sm = sm.add_constant(X_train)  # 상수항 추가
model_sm = sm.OLS(y_train, X_train_sm).fit()
print(model_sm.summary())
````

모델에서 유의하지 않은 설명변수는 [id, f3, f5] 총 3개이다.

````python
# 유의하지 않은 변수 제거 후 모델 재적합
X_train_reduced = X_train.drop(["id", "f3", "f5"], axis=1) 

# 모델 선언(다중 회귀)
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# 모델 학습
model.fit(X_train_reduced, y_train)

# 모델 예측 - 회귀계수 유의성 검증 후에 해달 칼럼 제거하여 예측해야하는 것 몰랐음
X_test_reduced = df3_1_test.drop(["id", "f3", "f5","design"], axis=1)
y_test_pred = model.predict(X_test_reduced)

# 예측값과 실제값 피어슨 상관계수(소수점 셋째 자리 반올림) - 몰랐음
from scipy.stats import pearsonr
y_test_actual = df3_1_test["design"]
pearson_corr, _ = pearsonr(y_test_actual, y_test_pred)
pearson_corr = round(pearson_corr, 3)
print(pearson_corr)     # 결과: 0.944
````

````python
# 모델 예측 점수 계산(RMSE)
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test_actual, y_test_pred)
print("RMSE:", rmse)        # RMSE: 19.149971930635683
````

### 2. 한 통신 회사에서는 고객 이탈을 줄이고자 주요 원인들을 분석하기로 결정하였다. 이를 위해 500명의 고객 데이터를 수집했으며, 고객의 서비스 이용 및 가입 정보, 그리고 일부 개인적인 속성을 조사하였다.

1. 고객 이탈을 예측하는 로지스틱 회귀를 시행한 후 col1 칼럼의 p-value를 구하시오 (소수점 넷째 자리에서 반올림)
2. 폰 서비스를 받지 않은 고객 대비 받은 고객의 이탈 확률 오즈비를 구하시오 (소수점 넷째 자리에서 반올림)
3. 이탈할 확률이 0.3이상인 고객 수를 구하시오.

````python
import pandas as pd

df3_2 = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_3_2.csv")
df3_2.head()
````

````python
df3_2.info()
````

````python
df3_2["Phone_Service"].value_counts()
````

````python
df3_2["Tech_Insurance"].value_counts()
````

````python
df3_2["churn"].value_counts()
````

````python
# 변수 분할
X_logistic = df3_2.drop("churn", axis=1)
y_logistic = df3_2["churn"]

# p-value 계산을 위한 statsmodels 사용(소수점 넷째 자리 반올림) - 몰랐음
import statsmodels.api as sm
X_logistic_sm = sm.add_constant(X_logistic)  # 상수항 추가
model_logistic_sm = sm.Logit(y_logistic, X_logistic_sm).fit()
print(model_logistic_sm.summary())
````

````python
# 로지스틱 회귀 모델 선언
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression(random_state=42)

model_logistic.fit(X_logistic, y_logistic)
````

````python
# 이탈 확률 오즈비 계산 - 몰랐음
import numpy as np
odds_ratios = pd.Series(model_logistic.coef_[0], index=X_logistic.columns)
odds_ratios = odds_ratios.apply(lambda x: round(np.exp(x), 4))
print(odds_ratios)
````

````python
# 이탈 확률이 0.3 이상인 고객 수 계산 - 몰랐음
y_pred_proba = model_logistic.predict_proba(X_logistic)[:, 1]
num_customers_high_risk = (y_pred_proba >= 0.3).sum()
print(num_customers_high_risk)  # 결과 449
````
